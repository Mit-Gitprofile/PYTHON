```python
import datetime
import csv
from bs4 import BeautifulSoup
import requests

events = []

try:
    with open("events.txt", "r") as file:
        events = [tuple(line.strip().split(", ")) for line in file]
except FileNotFoundError:
    pass

while True:
    print("\nEvent Planner Menu:")
    print("1. Add Event")
    print("2. Remove Event")
    print("3. View Events")
    print("4. Search Event")
    print("5. Save Events to File")
    print("6. Scrape Events")
    print("7. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        title = input("Enter event title: ")
        date = input("Enter event date (YYYY-MM-DD): ")
        time = input("Enter event time (HH:MM): ")
        location = input("Enter event location: ")
        events.append((title, date, time, location))

    elif choice == "2":
        title = input("Enter event title to remove: ")
        events = [event for event in events if event[0] != title]

    elif choice == "3":
        for event in events:
            print(event)

    elif choice == "4":
        keyword = input("Enter search keyword: ")
        results = [event for event in events if keyword.lower() in event[0].lower() or keyword.lower() in event[3].lower()]
        for event in results:
            print(event)

    elif choice == "5":
        with open("events.txt", "w") as file:
            for event in events:
                file.write(f"{event[0]}, {event[1]}, {event[2]}, {event[3]}\n")
        print("Events saved successfully.")

    elif choice == "6":
        url = "https://venngage.com/blog/templates-event-poster/"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        print(response.status_code)

        scraped_events = []
        for event_div in soup.find_all("h2", class_="wp-block-heading"):
            title = event_div.text
            scraped_events.append((title))
        
        for i in scraped_events:
            print(i)

        print("Events scraped and added successfully.")

    elif choice == "7":
        break
    else:
        print("Invalid choice. Please try again.")
```
